## 1. Load Libraries

Import the helper functions and any additional packages needed for analysis.

```python
from pathlib import Path

import pandas as pd
from src.helpers import (
    load_csv,
    preview_dataframe,
    save_csv,
    load_buildings_from_input,
)
from src.vizualizer import show_building

# Optional: define common paths for the project
DATA_DIR = Path('input')
OUTPUT_DIR = Path('output')
OUTPUT_DIR.mkdir(exist_ok=True)
```

## 2. Custom Inputs

Define project-specific paths and parameters that drive the notebook behavior.

```python
# Custom paths and parameters
INPUT_DIR = DATA_DIR
OUTPUT_DIR = OUTPUT_DIR
BUILDING_LOD = None
PREVIEW_ROWS = 10

print(f'Input directory: {INPUT_DIR}')
print(f'Output directory: {OUTPUT_DIR}')
print(f"Building LOD: {BUILDING_LOD or 'auto'}")
print(f'Preview rows: {PREVIEW_ROWS}')
```

## 3. Load Inputs

Load the input dataset from the `input/` folder and inspect the first rows.

```python
json_files = list(INPUT_DIR.rglob('lod3.json'))
print(f'Found {len(json_files)} lod3.json files in nested input folders')

buildings = load_buildings_from_input(INPUT_DIR, lod=BUILDING_LOD)
print(f'Loaded {len(buildings)} buildings')

if buildings:
    for building in buildings[:5]:
        print(f"- {building['name']} ({building['id']}): {len(building['geometry']['surfaces'])} surfaces")
else:
    print('No building geometries were found in the input JSON files.')
```

## 4. Core Functionality

Use helper functions to preview data and perform processing steps.

This section can be extended with custom cleaning, transformation, or analysis logic.

```python
if buildings:
    print('Preview of the first building:')
    first_building = buildings[0]
    print(f"Name: {first_building['name']}")
    print(f"Surface count: {len(first_building['geometry']['surfaces'])}")

    show_building(first_building)
else:
    print('No building data available for visualization.')
```

## 5. Output

Save processed results or preview data to the `output/` folder.

```python
import json

summary_path = OUTPUT_DIR / 'building_summary.json'
if buildings:
    summary = [
        {
            'id': building['id'],
            'name': building['name'],
            'surface_count': len(building['geometry']['surfaces']),
        }
        for building in buildings
    ]
    with open(summary_path, 'w', encoding='utf-8') as handle:
        json.dump(summary, handle, indent=2)
    print(f'Saved building summary to {summary_path}')
else:
    print('Skipping output save because no building data was loaded.')
```
